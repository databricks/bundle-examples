This is a helper notebook to update the column comments for all monitoring tables. This is a workaround for the current limitation in SDP where it is not possible to set the column comments for streaming tables and materialized views.

```python
import logging
import sys

sys.path.append("../lib")

from dbx_ingestion_monitoring.standard_tables import *
from dbx_ingestion_monitoring.common import *

dbutils.widgets.text("monitoring_catalog", "")
dbutils.widgets.text("monitoring_schema", "")

logging.basicConfig(level=logging.INFO, format="%(asctime)s [%(levelname)s] (%(name)s) %(message)s")
logging.getLogger("dbx_ingestion_monitoring.MonitoringTable").setLevel(logging.DEBUG)

monitoring_catalog = get_required_widget_parameter(dbutils.widgets, 'monitoring_catalog')
monitoring_schema = get_required_widget_parameter(dbutils.widgets, 'monitoring_schema')
set_all_table_column_comments(monitoring_catalog, monitoring_schema, spark)
```
